```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler  
```

```python
df=pd.read_csv("Clustering_gmm (2).csv")
df.head()
```

```python
print("Rows:",df.shape[0])
print("Columns:",df.shape[1])
```

```python
df.isnull().sum()
```

```python
df.duplicated().sum()
```

```python
X=df[['Weight','Height']]
X.head()
```

```python
plt.figure(figsize=(8,6))
plt.scatter(
    X['Weight'],
    X['Height'],
    color='blue'
)
```

```python
scalar = StandardScaler()
X_scaled = scalar.fit_transform(X)
```

```python
# ==========================================
# Step 13: Elbow Method
# ==========================================

wcss = []

for i in range(1,11):

    kmeans = KMeans(
        n_clusters=i,
        random_state=42
    )

    kmeans.fit(X_scaled)

    wcss.append(kmeans.inertia_)

plt.figure(figsize=(8,6))

plt.plot(
    range(1,11),
    wcss,
    marker='o'
)

plt.title("Elbow Method")

plt.xlabel("Number of Clusters")

plt.ylabel("WCSS")

plt.grid(True)

plt.show()
```
